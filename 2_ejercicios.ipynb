```python
# ejercicio 1
#Dado el arreglo a = np.array([1, 2, 3, 4]), determina si ninguno de los elementos es cero.
import numpy as np
a = np.array([1,2,3,4])
b = np.any(a==0)
if b == True:
    print("El arreglo contiene 0")
else:
    print("El arreglo no contiene 0")
```

```python
# ejercicio 2
#Dado el arreglo a = np.array([0, 0, 2, 0]), determina si contiene algún elemento que no sea cero.
a = np.array([0,0,2,0])
b = np.any(a!=0)
if b == True:
    print("El arreglo contiene elementos que no son 0")
else:
    print("El arreglo solo contiene 0")
```

```python
# ejercicio 3
#Analiza el arreglo a = np.array([1, 2, np.nan, np.inf]) y determina si contiene valores que sean infinitos o valores no numéricos (NaN).
a = np.array([1, 2, np.nan, np.inf])
b = np.isnan(a)
c = np.isinf(a)
d = np.any(b == True)
e = np.any(c == True)
if d == True:
    print("El arreglo contiene valores no numericos")
else:
    print("El arreglo solo contiene valores numericos")
if e == True:
    print("El arreglo contiene valores infinitos")
else:
    print("El arreglo solo contiene valores numericos")
```
